## Modifying Pixels

```python
duck_copy = duck.copy()
```

```python
duck_copy[0:20, 0:20] = 200
```

```python
plt.imshow(duck_copy, cmap='gray')
```

## Cropping Images
Select a specific pixel region of the image

```python
sunset_bgr = cv2.imread('./img/sunset_strip.jpg')
sunset_rgb = sunset_bgr[:,:,::-1]

plt.imshow(sunset_rgb)
```

```python
cropped_region = sunset_rgb[2200:2500, 2300:2800]
plt.imshow(cropped_region)
```

## Resizing Images
`resiez()`
arguments:
1. `src`: input image
2. `dsize`: output image size
3. `fx` (optional): Scale factor laong the horizontal axis
4. `fy` (optional): Scale factor along the vertical axis

```python
# 1. Specifying the Scaling factors fx and fy
resize_2x_cropped_region = cv2.resize(cropped_region, None, fx=2, fy=2)
plt.imshow(resize_2x_cropped_region)
```

```python
# 2. Specifying the exact size of the output
width = 200
height = 400
dim = (width, height)

resize_dim_cropped_region = cv2.resize(cropped_region, dsize=dim, interpolation=cv2.INTER_AREA)
plt.imshow(resize_dim_cropped_region)
```
